```python
from bs4 import BeautifulSoup
import requests
import re
```

```python
res = requests.get('https://news.yahoo.co.jp/list/?c=world')
```

```python
soup = BeautifulSoup(res.content, 'lxml')
```

```python
news_list = soup.find(class_='list')
news_list
```

```python
urls = []
for news in news_list:
    if news.find('a') != -1:
        urls.append(news.find('a').attrs['href'])
```

```python
resource = []
for url in urls:
    res = requests.get(url)
    soup = BeautifulSoup(res.content, 'lxml')
    source = soup.find(class_='newsLink').attrs['href']
    resource.append(source)
```

```python
for source in resource:
    res = requests.get(source)
    soup = BeautifulSoup(res.content, 'lxml')
    if soup.find(class_='ynDetailText'):
        text = soup.find(class_='ynDetailText').get_text()
        print(text)
```
